# 3. Adding a New Column
Example: Add a new column price_per_sqft calculated as price / total_sqft

---



```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Bengaluru_House_Data.csv')

# Function to clean and convert total_sqft values
def convert_sqft(value):
    value = str(value).strip()
    if '-' in value:
        # Take the average of the range
        parts = value.split('-')
        return (float(parts[0]) + float(parts[1])) / 2
    elif value.replace('.', '', 1).isdigit():
        # If it's a number, convert to float
        return float(value)
    else:
        # Handle cases with 'Sq. Meter', 'Acres', etc.
        numeric_part = ''.join(filter(str.isdigit, value))
        return float(numeric_part) if numeric_part else None

# Apply the function to the total_sqft column
df['total_sqft'] = df['total_sqft'].apply(convert_sqft)

# Drop rows where conversion failed
df.dropna(subset=['total_sqft'], inplace=True)

# Convert price column to numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Add new column: price per sqft
df['price_per_sqft'] = df['price'] / df['total_sqft']

print(df.head())
```

# Pivoting using pivot function

```python
import pandas as pd

# Sample data
data = {
    'area_type': ['Super built-up  Area', 'Plot  Area', 'Built-up  Area', 'Plot  Area'],
    'location': ['Whitefield', 'Indiranagar', 'Marathahalli', 'Electronic City'],
    'total_sqft': [1200, 1500, 1100, 1600],
    'bath': [2, 3, 2, 4],
    'price': [6000000, 7500000, 5000000, 8000000]
}

df = pd.DataFrame(data)

# Pivot table: Set location as index and area_type as columns, filling values with price
df_pivot = df.pivot(index='location', columns='area_type', values='price')

print(df_pivot)
```

```python
# Melting the columns 'price' and 'total_sqft'
df_melted = df.melt(id_vars=['location', 'area_type'], value_vars=['price', 'total_sqft'],
                    var_name='Metric', value_name='Value')

print(df_melted)
```
